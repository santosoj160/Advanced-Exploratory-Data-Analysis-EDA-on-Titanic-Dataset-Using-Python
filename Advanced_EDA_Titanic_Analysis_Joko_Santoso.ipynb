# Crosstab

```python
df.head()
```

```python
df.duplicated().sum()
```

```python
data = pd.crosstab(index = df['Pclass'], columns=df['Survived'], normalize='index')
```

```python
sns.heatmap(
    data,
    annot=True,
    fmt='.2%',
    cmap='viridis',
    cbar_kws={'label' : 'Percentage Survived'}
)
```

#Interpretasi:
"Grafik heatmap menunjukkan adanya korelasi kuat antara Pclass dan Survived. Penumpang Kelas 1 memiliki tingkat keselamatan tertinggi (62.96%), sedangkan penumpang Kelas 3 mengalami tingkat kematian tertinggi (75.76%). Hal ini mengindikasikan adanya pengaruh status sosial atau lokasi kabin terhadap peluang bertahan hidup penumpang Titanic."

```python
data = pd.crosstab(index = df['Sex'], columns=df['Survived'], normalize='index')
sns.heatmap(
    data,
    annot=True,
    fmt='.2f',
    cmap='viridis',
    cbar_kws={'label' : 'Percentage'}
)
```

#Interpretasi
Grafik heatmap ini menunjukkan bahwa jenis kelamin (Sex) merupakan faktor krusial dalam tingkat kelangsungan hidup penumpang. Mayoritas penumpang perempuan berhasil selamat (74%), sedangkan sebagian besar penumpang laki-laki meninggal dunia (81%). Ketimpangan yang kontras ini memperkuat bukti adanya prioritas evakuasi bagi perempuan selama tragedi berlangsung."

```python
data = pd.crosstab(index = df['Sex'], columns=df['Survived'], values=df['Age'], aggfunc='mean')
sns.heatmap(
    data,
    annot=True,
    fmt='.2f',
    cmap='viridis',
    cbar_kws={'label' : 'Average Age'}
)
```

#Interpretasi
"Grafik heatmap ini menunjukkan rata-rata umur berdasarkan jenis kelamin dan status keselamatan. Insight menariknya adalah pria yang meninggal memiliki rata-rata usia tertua (31.62 tahun), sedangkan pria yang selamat rata-rata berusia lebih muda (27.28 tahun). Sebaliknya, wanita yang selamat justru memiliki rata-rata usia yang sedikit lebih tua (28.85 tahun) dibandingkan wanita yang meninggal (25.05 tahun)."

```python
data = pd.crosstab(index = df['Sex'], columns=df['Survived'])
sns.heatmap(
    data,
    annot=True,
    fmt='.2f',
    cmap='viridis',
    cbar_kws={'label' : 'Count'}
)
```

```python
data = pd.pivot_table(data = df, index = 'Sex', columns='Survived', values='PassengerId',aggfunc='count')
```

```python
sns.heatmap(
    data,
    annot=True,
    fmt='.2f',
    cmap='viridis',
    cbar_kws={'label' : 'Count'}
)
```

#Interpretasi
"Grafik heatmap menunjukkan jumlah absolut penumpang berdasarkan jenis kelamin dan status keselamatan. Korban meninggal didominasi secara ekstrem oleh penumpang laki-laki yaitu sebanyak 468 orang (ditunjukkan oleh warna kuning cerah). Sebaliknya, jumlah penyintas (selamat) didominasi oleh perempuan sebanyak 233 orang."

## Group by Aggregation

```python
df.head()
```

```python
data = df.groupby('Sex')['Age'].mean().reset_index()
fig, ax = plt.subplots(figsize=(6, 4))
sns.barplot(data=data, x='Sex', y='Age', ax=ax)
```

#Interpretasi
"Grafik bar plot ini menunjukkan perbandingan rata-rata umur berdasarkan jenis kelamin penumpang. Dari hasil agregasi, dapat disimpulkan bahwa penumpang laki-laki memiliki karakteristik usia yang sedikit lebih tua (rata-rata ~30.7 tahun) dibandingkan dengan penumpang perempuan (rata-rata ~27.9 tahun). Secara keseluruhan, demografi penumpang Titanic didominasi oleh kelompok usia dewasa muda."
